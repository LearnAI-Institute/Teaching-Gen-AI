##### Connect with Google Gemini model

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
```

```python
res = llm.invoke("What is GenAI ? Explain me in short ?")
res
```

```python
print(res.content)
```

##### Connect with OpenAI model

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model="gpt-4o")
```

```python
res = llm.invoke("How to find largest number from a list in python ?")
print(res.content)
```

#### Static Prompts

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model="gpt-4o")
```

```python
prompts = [
    ("system", "You are a java developer."),    ### "system", "ai", , 'user'
    ("user", "How to sort the array.")
]

res = llm.invoke(prompts)
print(res.content)
```
