## 📝 Example 3: Query Parameters

**File:** `FastAPI_Starter_Examples/03_search.py`

**To run it:**
```bash
uvicorn 03_search:app --reload
# Visit: http://localhost:8000/search?q=python
# Visit: http://localhost:8000/search?q=fastapi
```

````python
# Code:
code_example_3 = """
from fastapi import FastAPI

app = FastAPI()

@app.get("/search")
def search(q: str):
    return {
        "you searched for": q,
        "results": ["result 1", "result 2"]
    }
"""

print(code_example_3)
print("\n🔗 Examples:")
print("  http://localhost:8000/search?q=python")
print("  http://localhost:8000/search?q=fastapi")
print("\nNote: Query comes AFTER the ?")
````

````python
# Test the logic
def search(q: str):
    results = {
        "python": ["tutorial", "docs", "forum"],
        "fastapi": ["docs", "tutorial", "github"],
        "django": ["framework", "tutorial", "docs"]
    }
    return {
        "search_term": q,
        "results": results.get(q, ["no results"])
    }

print("Testing:")
print(f"search('python') = {search('python')}")
print(f"\nsearch('fastapi') = {search('fastapi')}")
````

## 📝 Example 4: POST Request (Send Data)

**File:** `FastAPI_Starter_Examples/04_create_user.py`

**To test it:**
```bash
uvicorn 04_create_user:app --reload
# Open: http://localhost:8000/docs
# Try the POST endpoint
```

````python
# Code:
code_example_4 = """
from fastapi import FastAPI
from pydantic import BaseModel

app = FastAPI()

class User(BaseModel):
    name: str
    age: int
    email: str

users = []

@app.post("/users")
def create_user(user: User):
    users.append(user)
    return {"message": f"User {user.name} created!"}

@app.get("/users")
def get_users():
    return {"users": users}
"""

print(code_example_4)
print("\n📝 What happens:")
print("  1. Define User model (what data looks like)")
print("  2. POST /users creates a user")
print("  3. GET /users shows all users")
print("\n⚠️  Data is validated automatically!")
````

````python
# Test the logic
from pydantic import BaseModel

class User(BaseModel):
    name: str
    age: int
    email: str

users = []

def create_user(user: User):
    users.append(user)
    return {"message": f"User {user.name} created!"}

# Create users
ali = User(name="Ali", age=25, email="ali@example.com")
fatima = User(name="Fatima", age=23, email="fatima@example.com")

print("Creating users:")
print(create_user(ali))
print(create_user(fatima))
print(f"\nTotal users: {len(users)}")
print(f"Users: {[u.name for u in users]}")
````

## 📝 Example 5: Calculator API (Multiple Operations)

**File:** `FastAPI_Starter_Examples/05_calculator.py`

**To run it:**
```bash
uvicorn 05_calculator:app --reload
# Visit: http://localhost:8000/add?a=5&b=3
# Visit: http://localhost:8000/multiply?a=4&b=7
```

````python
# Code structure:
code_example_5 = """
from fastapi import FastAPI

app = FastAPI()

@app.get("/add")
def add(a: int, b: int):
    return {"result": a + b, "calculation": f\"{a} + {b}\"}

@app.get("/multiply")
def multiply(a: int, b: int):
    return {"result": a * b, "calculation": f\"{a} * {b}\"}

@app.get("/divide")
def divide(a: int, b: int):
    if b == 0:
        return {"error": "Cannot divide by zero!"}
    return {"result": a / b, "calculation": f\"{a} / {b}\"}
"""

print(code_example_5)
print("\n🔗 Examples:")
print("  http://localhost:8000/add?a=5&b=3")
print("  http://localhost:8000/multiply?a=4&b=7")
print("  http://localhost:8000/divide?a=20&b=0  (error handling!)")
````

````python
# Test the logic
def add(a: int, b: int):
    return {"a": a, "b": b, "result": a + b}

def multiply(a: int, b: int):
    return {"a": a, "b": b, "result": a * b}

def divide(a: int, b: int):
    if b == 0:
        return {"error": "Cannot divide by zero!"}
    return {"a": a, "b": b, "result": a / b}

print("Testing:")
print(f"add(5, 3) = {add(5, 3)}")
print(f"\nmultiply(4, 7) = {multiply(4, 7)}")
print(f"\ndivide(20, 4) = {divide(20, 4)}")
print(f"\ndivide(20, 0) = {divide(20, 0)}  ← Error handling!")
````

## 🎯 Key Concepts (Just the Code)

### 1️⃣ GET Request (Read Data)
```python
@app.get("/path")
def my_function():
    return {"data": "response"}
```

### 2️⃣ Path Parameter (Variable in URL)
```python
@app.get("/user/{user_id}")
def get_user(user_id: int):
    return {"id": user_id}
```

### 3️⃣ Query Parameter (After ?)
```python
@app.get("/search")
def search(q: str):
    return {"search": q}
```

### 4️⃣ POST Request (Send Data)
```python
from pydantic import BaseModel

class Item(BaseModel):
    name: str
    price: float

@app.post("/items")
def create_item(item: Item):
    return {"created": item}
```

### 5️⃣ Data Validation (Automatic!)
```python
def function(age: int, email: str):
    # FastAPI checks that age is int and email is str
    # If wrong type → automatic error
```

## ✅ Difference: Path vs Query Parameters

**PATH PARAMETER** (in the URL path):
```python
@app.get("/user/{user_id}")
def get_user(user_id: int):
    ...

# URL: /user/123
```

**QUERY PARAMETER** (after the ?):
```python
@app.get("/search")
def search(q: str, category: str):
    ...

# URL: /search?q=python&category=tutorial
```

## ⚠️ Common Mistakes

### ❌ Mistake 1: Port already in use
```bash
# Error: Port 8000 already in use
# Solution: Use different port
uvicorn main:app --port 8001 --reload
```

### ❌ Mistake 2: Wrong data type
```python
# Code expects int, user sends string
@app.get("/divide")
def divide(a: int, b: int):
    return {"result": a / b}

# Wrong: /divide?a=20&b=zero  ← "zero" is not int
# Right: /divide?a=20&b=4     ← both are numbers
```

### ❌ Mistake 3: Forgetting {}
```python
# Wrong:
@app.get("/user/name")
def get_user(name):
    ...

# Right:
@app.get("/user/{name}")
def get_user(name: str):
    ...
```
